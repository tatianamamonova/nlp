```python
import nltk
import re
```

```python
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

Импортируем большой текст и чистим его

```python
with open('text_1.txt', 'r') as f:
    t = f.read()
```

```python
t
```

```python
def clear_text(te):
    clear_te = re.sub('\\xa0', ' ', te)
    clear_te = re.sub('\n\n', '\n', clear_te)
    return clear_te
```

```python
text = clear_text(t)
```

```python
with open('text.txt', 'r') as f:
    te = f.read()
```

```python
clear_te = clear_text(te)
```

```python
clear_te
```

```python
with open('text_splitted.txt', 'r') as f:
    t_spl = f.read()

clear_text_ideal = clear_text(t_spl)
ideal = clear_text_ideal.split('\n')
```

```python
for i, line in enumerate(ideal):
    ideal[i] = line.strip(' ')
ideal
```

Делим на предложения маленький текст

```python
re.split('(?<=[.!?]) (?=[А-Я—])|\n(?=[А-Я—])', clear_te)
```

```python
tp = 0
fp = 0
fn = 0

for sent in ideal:
    if len(re.split('(?<=[.!?]) (?=[А-Я—])|\n(?=[А-Я—])', sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(ideal)-1):
    sent1, sent2 = ideal[i], ideal[i+1]
    sent = ' '.join([sent1, sent2])
    if len(re.split('(?<=[.!?]) (?=[А-Я—])|\n(?=[А-Я—])', sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(text)
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
tokenizer.tokenize(clear_te)
```

```python
tp = 0
fp = 0
fn = 0

for sent in ideal:
    if len(tokenizer.tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(ideal)-1):
    sent1, sent2 = ideal[i], ideal[i+1]
    sent = ' '.join([sent1, sent2])
    if len(tokenizer.tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
